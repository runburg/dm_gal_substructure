```python
r, th, ph = symbols(r'r \theta \phi')
rho, drho, P, dP, lam = symbols(r'\rho \delta_\rho P \delta_P \lambda', cls=Function)
```

```python
u = Matrix([[1, 0, 0, 0], [0, sin(th)*cos(ph), cos(th)*cos(ph), -sin(ph)], [0, sin(th)*sin(ph), cos(th)*sin(ph), cos(ph)], [0, cos(th), -sin(th), 0]])
u
```

```python
T = Matrix([[-rho(r) - drho(r), 0, 0, 0], [0, P(r) + dP(r), 0, 0], [0, 0, P(r) + dP(r) + lam(r), 0], [0, 0, 0, P(r) + dP(r) + lam(r)]])
T
```

```python
res = simplify(u*T*(u.T))
res
```

```python
res = res.subs(th, acos(z/r))
res = res.subs(ph, atan(y/x))
res
```
